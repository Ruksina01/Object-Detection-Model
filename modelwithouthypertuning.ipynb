# Output - 1

```python
import pandas as pd
import cv2
import matplotlib.pyplot as plt
log_file_path = "/kaggle/working/runs/detect/train2/results.csv"
logs = pd.read_csv(log_file_path)  # Read log file into a DataFrame

final_losses = logs.iloc[-1]  # Get the last row

# Print final losses
print("Final Losses:")
print(f"Training Box Loss: {final_losses['train/box_loss']}")
print(f"Training Class Loss: {final_losses['train/cls_loss']}")
print(f"Training DFL Loss: {final_losses['train/dfl_loss']}")
print(f"Validation Box Loss: {final_losses['val/box_loss']}")
print(f"Validation Class Loss: {final_losses['val/cls_loss']}")
print(f"Validation DFL Loss: {final_losses['val/dfl_loss']}")
```

```python
plt.figure(figsize=(12, 5))

# Plot training box loss
plt.subplot(1, 2, 1)
plt.plot(logs['epoch'], logs['train/box_loss'], label='Training Box Loss', color='blue')
plt.plot(logs['epoch'], logs['val/box_loss'], label='Validation Box Loss', color='red')
plt.title('Training and Validation Box Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.grid()
plt.legend()
```

```python
path="/kaggle/input/pesmodsplitdataset/test/images/frame0137.jpg"
```

```python
#prediction on a single image
results = model.predict(source=path)
```

```python
import cv2
image = cv2.imread(path)

for result in results:
    boxes = result.boxes  # Get the boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0].int().tolist()  # Get box coordinates
        conf = box.conf[0].item()  # Get confidence score
        cls = box.cls[0].item()  # Get class id
        label = f'{model.names[int(cls)]} {conf:.2f}'  # Create label with confidence

        # Draw bounding box and label
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        #cv2.putText(image, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Convert the image from BGR to RGB for displaying
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using matplotlib
plt.title("Objects: " + str(len(boxes)))
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
